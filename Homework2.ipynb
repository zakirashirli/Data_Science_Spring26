```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Q1-Read the data https://raw.githubusercontent.com/ogut77/DataScience/main/data/Credit.csv
url = "https://raw.githubusercontent.com/ogut77/DataScience/main/data/Credit.csv"

df = pd.read_csv(url)
```

```python
#Q2- Get data information and number of rows and columns
df.info()
df.shape
rows = df.shape[0]
columns = df.shape[1]

print("Number of rows:", rows)
print("Number of columns:", columns)

```

```python
#Q3-Get first 3 rows of data
df.head(3)
```

```python
#Q4-Get descriptive statistics of data
df.describe()
```

```python
#Q5-Use seasoborns heat map to generate following heatmap
# annot = True to print the correlation inside the square
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
#Q6-Plot the density distribution of Balance
sns.kdeplot(df["Balance"], fill=True)
plt.title("Density Distribution of Balance")
plt.xlabel("Balance")
plt.ylabel("Density")
plt.show()
```

```python
#Q7-Count the frequency values of each Region
df["Region"].value_counts()
```

```python
#Q8-List Percentage of  the Married people
married_percentage = df["Married"].value_counts(normalize=True) * 100
married_percentage
```

```python
#Q9-Create histogram of Balance
plt.hist(df["Balance"], bins=20)
plt.title("Histogram of Balance")
plt.xlabel("Balance")
plt.ylabel("Frequency")
plt.show()
```

```python
#Q10-Get the frequency distribution of Region using bar graph
region_counts = df["Region"].value_counts()

region_counts.plot(kind="bar")
plt.title("Frequency Distribution of Region")
plt.xlabel("Region")
plt.ylabel("Frequency")
plt.show()
```

```python
#Q11-Get the mean income  of married people?
mean_income_married = df[df["Married"] == "Yes"]["Income"].mean()
mean_income_married
```

```python
#Q12-Get the scatter plot of Income and Limit
plt.scatter(df["Income"], df["Limit"])
plt.title("Scatter Plot of Income vs Limit")
plt.xlabel("Income")
plt.ylabel("Limit")
plt.show()
```

```python
#Q13-Create new column named IncomebyLimit by dividing Income to Limit.
df["IncomebyLimit"] = df["Income"] / df["Limit"]
df.head()
```

```python
# Q14-Create the pair plot using Seaborn library's pairplot
sns.pairplot(df)
plt.show()
```

```python
#Q15- drop new column named IncomebyLimit from dataframe
df = df.drop("IncomebyLimit", axis=1)
df.head()
```
