```python
#Q8-Write a class named arithmetic with method for calculating addition, substraction, multiplication and division of two numbers.
class arithmetic:
  def addition(x,y):
    return x+y

  def substraction(x,y):
    return x - y

  def multiplication(x,y):
    return x * y

  def division(x,y):
    if y == 0:
      return 'You cannot divide by zero'
    else:
      return x / y

print('addition:',arithmetic.addition(3,4))
print('substraction:', arithmetic.substraction(3,4))
print('multiplication:',arithmetic.multiplication(3,4))
print('division:', arithmetic.division(12,0))
```

```python
#Q9- write a for loop or method  to calculate summation of first n numbers
#i.e if n=3, it will print or return 6 (1+2+3)
n = int(input("Enter a number: "))
total = 0
for i in range(1, n + 1):
  total += i
  print(total)
```
